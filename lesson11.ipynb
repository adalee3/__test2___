```python
import numpy as np
np.linspace(-3,3) #引數值呼叫
```

```python
array=np.linspace(-3,3,num=100) #num 引數名稱呼叫
array*10
```

```python
np.random.randint(50,high=101,size=10) #high  不包含 tuple裡面不可修改，用小括號,引數值要依順序，引數名稱可不依序
```

```python
np.random.randint(50,high=101,size=(50,5)) #tuple,二維表格陣列，50個學生,5個分數
```
